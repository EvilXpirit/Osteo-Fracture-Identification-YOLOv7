```python
from data_loader import XrayDataset
import numpy as np
import matplotlib.pyplot as plt

annotations_file = "GRAZPEDWRI-DX_dataset/test_data.csv"
root = "GRAZPEDWRI-DX_dataset"
pred_root = "runs/test/eval_test/labels"

ds = XrayDataset(annotations_file, root)
ds.describe()
```

```python
N = 6
rows = 1
rnd_idxs = np.random.choice(np.arange(len(ds)), size=N, replace=False)

plt.figure(figsize=(15,10))
for i, idx in enumerate(rnd_idxs):
    plt.subplot(rows, N//rows, i+1)
    plt.imshow(ds.blend_data(idx))
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
ds2 = XrayDataset(annotations_file, img_root=root, pred_root=pred_root, conf_thres=0.4)

plt.figure(figsize=(15,10))
for i, idx in enumerate(rnd_idxs):
    plt.subplot(rows, N//rows, i+1)
    plt.imshow(ds2.blend_data(idx))
    plt.axis("off")
plt.tight_layout()
plt.show()
```
